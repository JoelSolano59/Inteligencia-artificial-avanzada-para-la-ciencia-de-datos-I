**Prueba Z**
Hipotesis nula
${_0:\mu_1=\mu_2}$
Estadistica de prueba
$Z=\frac{\overline{x}-\overline{y}-(\mu_1-\mu_2)}{\sqrt{\frac{\sigma_1^2}{m}+\frac{\sigma_2^2}{n}}}$
El estadistico de prueba tiene una distribucion normal estandar

```python
#Realizar el z-test
from statsmodels.stats.weightstats import ztest as ztest
ztest(df['Cliente A'], df['Cliente B'])
```

##Prueba T
**Distribucion t-student**

```python
!pip install pingouin
import pingouin as pg
# Descarga de datos
# ==============================================================================
url = ('https://raw.githubusercontent.com/JoaquinAmatRodrigo/' +
       'Estadistica-machine-learning-python/master/data/births.csv')
datos = pd.read_csv(url, sep=',')
datos.head()
```

**Prueba T-student**
Hipotesis nula
${H_0: \mu_1 = \mu_2}$
Estadistico de prueba
$t=\frac{\overline{x}-\overline{y}-(\mu_1-\mu_2)}{\sqrt{\frac{\S_1^2}{m}+\frac{\S_2^2}{n}}}$
El estadistico de prueba tiene una distribucion t-student

```python
# Filtrar la base de datos en madres fumadoras y no fumadoras
peso_smokers = datos.loc[datos.smoke == 'smoker', 'weight']
peso_nonsmokers = datos.loc[datos.smoke == 'nonsmoker', 'weight']
peso_smokers
peso_nonsmokers
```

```python
#aplicando el t-test
pg.ttest(x = peso_smokers, y = peso_nonsmokers)
```
